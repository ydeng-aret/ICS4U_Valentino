```python
import math
class quadratic_func:
    def __init__(self, a, b, c, h, k):
        self.a = a
        self.b = b
        self.c = c
        self.h = h
        self.k = k
        
    def standard_form(self):
        return f"The standard form is {self.a}x^2+({self.b})x+{self.c}"
        
    def vertex_form(self):
        return f"The standard form is {self.a}(x-({self.h}))^2+{self.k}"

    def concavity(self):
        if self.a > 0:
            return "The graph opens up"
        else:
            return "The graph opens down"

    def vertex_coordinates(self):
        return f"The vertex is ({self.h},{self.k})"
    
    def y_intercept(self):
        return f"The y-intercept is y={self.c}"

    def x_intercepts(self):
        discriminant = self.b ** 2 - 4 * self.a * self.c     #Set the discriminant to see whether it is positive, negative, or zero
        if discriminant < 0:
            return []
        elif discriminant == 0:
            x = -self.b / (2 * self.a)
            return [x]
        else:
            sqrt_d = math.sqrt(discriminant)
            x1 = (-self.b + sqrt_d) / (2 * self.a)
            x2 = (-self.b - sqrt_d) / (2 * self.a)
            return [x1, x2]

    
    def add(self, other):
        new_a = self.a + other.a
        new_b = self.b + other.b
        new_c = self.c + other.c
        new_h = self.h + other.h
        new_k = self.k + other.k
        return quadratic_func(new_a, new_b, new_c, new_h, new_k)
    

    

a = float(input("Please type the value of A for the first function: "))
b = float(input("Please type the value of B for the first function: "))
c = float(input("Please type the value of C for the first function: "))
h = float(b/(-2 * a))
k = a * h ** 2 + b * h + c

a2 = float(input("Please type the value of A for the second function: "))
b2 = float(input("Please type the value of B for the second function: "))
c2 = float(input("Please type the value of C for the second function: "))
h2 = float(b2/(-2 * a2))
k2 = a2 * h2 ** 2 + b2 * h2 + c2

def main():
    print("---- Quadratic 1 ----")
    q1 = quadratic_func(a, b, c, h, k)
    print("Standard Form:", q1.standard_form())
    print("Vertex Form:", q1.vertex_form())
    print("Direction:", q1.concavity())
    print("Vertex:", q1.vertex_coordinates())
    print("Y-Intercept:", q1.y_intercept())
    print("X-Intercepts:", q1.x_intercepts())

    q2 = quadratic_func(a2, b2, c2, h2, k2)
    print("Standard Form:", q2.standard_form())
    print("Vertex Form:", q2.vertex_form())
    print("Direction:", q2.concavity())
    print("Vertex:", q2.vertex_coordinates())
    print("Y-Intercept:", q2.y_intercept())
    print("X-Intercepts:", q2.x_intercepts())


    print("\n---- Sum of Quadratic 1 and 2 ----")
    q3 = q1.add(q2)
    print("Standard Form:", q3.standard_form())
    print("Vertex Form:", q3.vertex_form())
    print("Direction:", q3.concavity())
    print("Vertex:", q3.vertex_coordinates())
    print("Y-Intercept:", q3.y_intercept())
    print("X-Intercepts:", q3.x_intercepts())

main()


```
